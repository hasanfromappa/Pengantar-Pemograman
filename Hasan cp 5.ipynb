```python
import pandas as pd # untuk membaca data dari file excel
from scipy.stats.mstats import gmean, hmean # untuk menghitung geometric mean dan harmonic mean
import matplotlib.pyplot as plt # untuk membuat histogram dan menampilkan garis vertikal untuk mean

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel

a_mean = my_dataset.Tm.mean() # menghitung arithmetic mean
g_mean = gmean(my_dataset['Tm'])        # menghitung geometric mean
h_mean = hmean(my_dataset['Tm']) # menghitung harmonic mean

print ('-------') # menampilkan hasil dengan format 1 angka di belakang koma
print ('arithmetic mean') # menampilkan hasil dengan format 1 angka di belakang koma
print ("{0:.1f} [ppm]".format(a_mean))  # menampilkan hasil dengan format 1 angka di belakang koma
print ('-------') # menampilkan hasil dengan format 1 angka di belakang koma

print ('geometric mean') # menampilkan hasil dengan format 1 angka di belakang koma
print ("{0:.1f} [ppm]".format(g_mean)) # menampilkan hasil dengan format 1 angka di belakang koma
print ('-------')

print ('harmonic mean') # menampilkan hasil dengan format 1 angka di belakang koma
print ("{0:.1f} [ppm]".format(h_mean)) # menampilkan hasil dengan format 1 angka di belakang koma
print ('-------')

fig, ax = plt.subplots() # membuat histogram dengan jumlah bin yang otomatis disesuaikan
ax.hist(my_dataset.Tm, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) # membuat histogram dengan jumlah bin yang otomatis disesuaikan
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2) # menambahkan garis vertikal untuk arithmetic mean
ax.axvline(g_mean, color='orange',  label='Geometric mean', linewidth=2)    # menambahkan garis vertikal untuk geometric mean
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=2) # menambahkan garis vertikal untuk harmonic mean
ax.set_xlabel('Tm [ppm]') # menambahkan label untuk sumbu x
ax.set_ylabel('Probability density') # menambahkan label untuk sumbu x dan y
ax.legend()

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel

variance =  my_dataset['Tm'].var() # menghitung variance dengan menggunakan method var() dari pandas yang secara otomatis akan menghitung rata-rata dari data, kemudian mencari selisih antara setiap data dengan rata-rata, mengkuadratkan selisih tersebut, dan mengambil rata-rata dari hasil kuadrat selisih tersebut
stddev =  my_dataset['Tm'].std() # menghitung standard deviation dengan menggunakan method std() dari pandas yang secara otomatis akan menghitung variance terlebih dahulu, kemudian mengambil akar kuadrat dari variance tersebut untuk mendapatkan standard deviation

print ('-------')
print ('Variance') # menampilkan hasil dengan format 1 angka di belakang koma
print("{0:.0f} [square ppm]".format(variance)) # menampilkan hasil dengan format 0 angka di belakang koma karena variance biasanya ditampilkan dalam bentuk bilangan bulat
print ('-------')
print ('Standard Deviation') # menampilkan hasil dengan format 1 angka di belakang koma
print("{0:.0f} [ppm]".format(stddev)) # menampilkan hasil dengan format 0 angka di belakang koma karena standard deviation biasanya ditampilkan dalam bentuk bilangan bulat
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Tm, bins= 20, density = True, edgecolor='k', label='Measurements Hist') # membuat histogram dengan jumlah bin yang otomatis disesuaikan
ax.axvline(my_dataset['Tm'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) # menambahkan garis vertikal untuk mean - 1 sigma
ax.axvline(my_dataset['Tm'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2) # menambahkan garis vertikal untuk mean + 1 sigma
ax.axvspan(my_dataset['Tm'].mean()-stddev, my_dataset['Tm'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') # menambahkan area berwarna untuk menunjukkan rentang antara mean - 1 sigma dan mean + 1 sigma
ax.set_xlabel('Tm [ppm]') # menambahkan label untuk sumbu x
ax.set_ylabel('Probability density') # menambahkan label untuk sumbu x dan y
ax.legend()

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np  # untuk menghitung kuartil dan interquartile range (IQR)
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel

q1 = np.percentile(my_dataset.Tm, 25, method = 'midpoint') # menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) menggunakan metode midpoint untuk menentukan nilai kuartil yang tepat ketika data memiliki jumlah elemen genap
q3 = np.percentile(my_dataset.Tm, 75, method = 'midpoint')  # menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) menggunakan metode midpoint untuk menentukan nilai kuartil yang tepat ketika data memiliki jumlah elemen genap
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) # menampilkan hasil dengan format 0 angka di belakang koma karena IQR biasanya ditampilkan dalam bentuk bilangan bulat
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Tm, bins='auto', density=True, edgecolor='k', label='Measurements Hist') # membuat histogram dengan jumlah bin yang otomatis disesuaikan dengan data
ax.axvline(q1, color='purple', label='Q1', linewidth=2) # menambahkan garis vertikal untuk kuartil pertama (Q1)
ax.axvline(q3, color='green', label='Q3', linewidth=2) # menambahkan garis vertikal untuk kuartil ketiga (Q3)
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') # menambahkan area berwarna untuk menunjukkan rentang antara Q1 dan Q3 yang merupakan interquartile range (IQR)
ax.set_xlabel('Tm [ppm]') # menambahkan label untuk sumbu x
ax.set_ylabel('Probability density') # menambahkan label untuk sumbu x dan y
ax.legend()

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel

a_mean = my_dataset.Tm.mean() # menghitung arithmetic mean dengan menggunakan method mean() dari pandas yang secara otomatis akan menjumlahkan semua data dan membaginya dengan jumlah data untuk mendapatkan nilai rata-rata

median = my_dataset.Tm.median() # menghitung median dengan menggunakan method median() dari pandas yang secara otomatis akan mengurutkan data dan mengambil nilai tengahnya

hist, bin_edges = np.histogram(my_dataset['Tm'], bins=20, density=True) # membuat histogram dengan menggunakan fungsi histogram() dari numpy yang akan menghasilkan array hist yang berisi frekuensi untuk setiap bin dan array bin_edges yang berisi batas bawah dan batas atas untuk setiap bin
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi dan mengambil rata-rata dari batas bawah dan batas atas bin tersebut

fig, ax = plt.subplots()
ax.hist(my_dataset.Tm, bins=20, density=True, edgecolor='k', label="Measurements Hist") # membuat histogram dengan jumlah bin yang otomatis disesuaikan
ax.axvline(modal_value, color='orange', label='Modal Value', linewidth=2) # menambahkan garis vertikal untuk modal value
ax.axvline(median, color='purple',  label='Median Value', linewidth=2) # menambahkan garis vertikal untuk median
ax.axvline(a_mean, color='green',  label='Arithmetic mean', linewidth=2) # menambahkan garis vertikal untuk arithmetic mean
ax.set_xlabel('Tm [ppm]') # menambahkan label untuk sumbu x
ax.set_ylabel('Probability density') # menambahkan label untuk sumbu x dan y
ax.legend()

```

```python
import numpy as np
from scipy.stats import skew # untuk menghitung skewness dengan menggunakan fungsi skew() dari scipy.stats yang secara otomatis akan menghitung skewness berdasarkan rumus yang sesuai dengan jenis data yang digunakan

a_mean = my_dataset.Tm.mean() # menghitung arithmetic mean
median = my_dataset.Tm.median()
hist, bin_edges = np.histogram(my_dataset['Tm'], bins=20, density=True) # membuat histogram dengan menggunakan fungsi histogram() dari numpy yang akan menghasilkan array hist yang berisi frekuensi untuk setiap bin dan array bin_edges yang berisi batas bawah dan batas atas untuk setiap bin
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 # menghitung nilai modal dengan mencari bin yang memiliki frekuensi tertinggi dan mengambil rata-rata dari batas bawah dan batas atas bin tersebut
standard_deviation = my_dataset['Tm'].std() # menghitung standard deviation dengan menggunakan method std() dari pandas yang secara otomatis akan menghitung variance terlebih dahulu, kemudian mengambil akar kuadrat dari variance tersebut untuk mendapatkan standard deviation

a1 = (a_mean - modal_value) / standard_deviation # menghitung Pearson's first coefficient of skewness dengan menggunakan rumus yang sesuai untuk data yang digunakan
a2 = 3 * (a_mean - median) / standard_deviation # menghitung Pearson's 2nd moment of skewness dengan menggunakan rumus yang sesuai untuk data yang digunakan
g1 = skew(my_dataset['Tm']) # menghitung Fisher-Pearson's coefficient of skewness dengan menggunakan fungsi skew() dari scipy.stats yang secara otomatis akan menghitung skewness berdasarkan rumus yang sesuai dengan jenis data yang digunakan

print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) # menampilkan hasil dengan format 2 angka di belakang koma karena skewness biasanya ditampilkan dalam bentuk bilangan desimal dengan beberapa angka di belakang koma
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) # menampilkan hasil dengan format 2 angka di belakang koma karena skewness biasanya ditampilkan dalam bentuk bilangan desimal dengan beberapa angka di belakang koma
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) # menampilkan hasil dengan format 2 angka di belakang koma karena skewness biasanya ditampilkan dalam bentuk bilangan desimal dengan beberapa angka di belakang koma
print ('-------')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
